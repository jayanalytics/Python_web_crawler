```python
import time
import requests
import numpy as np

from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait



#initialize variables
txtdata =time.strftime("%d/%b/%Y %H:%M:%S")
print("Started - " + txtdata)
account = "C4"


souq = []
str1 = "Model|Price|Date Extracted|Retailer"
souq.append (str1)

#Specify path where phantonjs.exe is present on your local system
browser = webdriver.PhantomJS(executable_path='......./phantomjs-2.1.1-windows/bin/phantomjs.exe')

#To get all Smart phones---------------------------------------------------------------------------------------------- 
page=0
morepage=1
while (page<=3):
    url="https://www.carrefouruae.com/mafuae/en/smartphones-tablets/mobile-phones/smartphones/c/NF8010100?&qsort=relevance&pg=" + str(page)
    
    #To get all phones below 499
    browser.get(url)
    browser.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    PhoneType= "Smart Phone"
    html = browser.page_source
    soup = BeautifulSoup(html, 'lxml')
    
    pdata = soup.find_all("div", {"class":"product"})

    for item in pdata:
        pdata1 = item.find("a")
        modelname = pdata1.get("title")

        pdata2= item.find("img")
        modelname1= pdata2.get("title")

        pdata3 = item.find("p",{"class":"finalPrice"})
        if pdata3 is not None:
            price = pdata3.text
        else:
            pdata3 = item.find("div",{"class":"price discount"})
            price = pdata3.find("del").text
        
        #send data to variable
        str1 = modelname + "|" + modelname1 + "|" + price + "|" + txtdata + "|" + account + "|" + PhoneType
        souq.append (str1)
    page=page+1
    
    
    #Check if last pages are done
    #<p class="facet-name">0 items</p>
    pdataTest = soup.find("p", {"class":"facet-name"})
    if pdataTest is not None:
        morepage='Pages done'
#--------------------------------------------------------------------------------------------------------

#To get all portable phones---------------------------------------------------------------------------------------------- 
page=1
morepage='yes'
while (morepage=='yes'):
    url="https://www.carrefouruae.com/mafuae/en/smartphones-tablets/mobile-phones/portable-phones/c/NF8010200"
    
    #Breaking the loop since there is only one page on protable phones in C4
    morepage='no'
    
    browser.get(url)
    browser.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(3.6)
    PhoneType= "Feature Phone"
    html = browser.page_source
    soup = BeautifulSoup(html, 'lxml')
    
    pdata = soup.find_all("div", {"class":"product"})

    for item in pdata:
        pdata1 = item.find("a")
        modelname = pdata1.get("title")
        modelname=modelname.strip()
        
        pdata2= item.find("img")
        modelname1= pdata2.get("title")
        modelname1=modelname1.strip()

        pdata3 = item.find("p",{"class":"finalPrice"})
        if pdata3 is not None:
            price = pdata3.text
            price = price.strip('\t')
        else:
            pdata3 = item.find("div",{"class":"price discount"})
            price = pdata3.find("del").text
            price = price.strip('\t')
        
        #send data to variable
        txtdata=txtdata.strip()
        account=account.strip()
        PhoneType=PhoneType.strip()
        str1 = modelname + "|" + modelname1 + "|" + price + "|" + txtdata + "|" + account
        souq.append (str1)
    page=page+1
    
#--------------------------------------------------------------------------------------------------------

browser.close()
browser.quit()

# Create an new Excel file and add a worksheet.
import xlsxwriter
txtdata1 =time.strftime("%d-%b-%Y %H%M%S")
import xlsxwriter
name1 = 'Data ' + txtdata1 + '.xlsx'
fileName = '...../DailyData/' + name1 #Specify path where data needs to be copied

workbook = xlsxwriter.Workbook(fileName)
worksheet = workbook.add_worksheet()

cell=1
for item in souq:
    address1 = 'A' + str(cell)
    worksheet.write(address1, item)
    cell=cell+1

workbook.close()

txtdata =time.strftime("%d/%b/%Y %H:%M:%S")
print("Ended - " + txtdata)
```
